```python
#(2)busca binária


import time
import numpy as np
from ast import Str
import string

f = open('br-utf8.txt', 'r', encoding='utf-8')
lista = f.readlines()
f.close()

array = np.empty(len(lista), dtype=Str)

for i in range(0, len(lista)):
  nome = lista[i]
  array[i] = nome[0:len(nome)-1]


def busca_binaria(array, nome, inicio = 0, fim = None):
    if fim is None:
        fim = len(array)-1
    if inicio <= fim:
        m = (inicio + fim) // 2
        if array[m] == nome:
            return m
        if nome < array[m]:
            return nome
        else:
            return busca_binaria(array, nome, m+1, fim)
    return None

a = busca_binaria(array, "não")
print(a)
%timeit busca_binaria(array, "não")
b = busca_binaria(array, "atire")
print(b)
%timeit busca_binaria(array, "atire")
c = busca_binaria(array, "o")
print(c)
%timeit busca_binaria(array, "o")
d = busca_binaria(array, "pau")
print(d)
%timeit busca_binaria(array, "pau")
e = busca_binaria(array, "no")
print(e)
%timeit busca_binaria(array, "no")
f = busca_binaria(array, "gato")
print(f)
%timeit busca_binaria(array, "gato")
g = busca_binaria(array, "porque")
print(g)
%timeit busca_binaria(array, "porque")
h = busca_binaria(array, "isso")
print(h)
%timeit busca_binaria(array, "isso")
i = busca_binaria(array, "não")
print(i)
%timeit busca_binaria(array, "não")
j = busca_binaria(array, "se")
print(j)
%timeit busca_binaria(array, "se")
k = busca_binaria(array, "faz")
print(k)
%timeit busca_binaria(array, "faz")
l = busca_binaria(array, "o")
print(l)
%timeit busca_binaria(array, "o")
m = busca_binaria(array, "gatinho")
print(m)
%timeit busca_binaria(array, "gatinho")
n = busca_binaria(array, "é")
print(n)
%timeit busca_binaria(array, "é")
o = busca_binaria(array, "nosso")
print(o)
%timeit busca_binaria(array, "nosso")
p = busca_binaria(array, "amigo")
print(p)
%timeit busca_binaria(array, "amigo")
q = busca_binaria(array, "não")
print(q)
%timeit busca_binaria(array, "não")
r = busca_binaria(array, "devemos")
print(r)
%timeit busca_binaria(array, "devemos")
s = busca_binaria(array, "maltratar")
print(s)
%timeit busca_binaria(array, "maltratar")
t = busca_binaria(array, "os")
print(t)
%timeit busca_binaria(array, "os")
u = busca_binaria(array, "animais")
print(u)
%timeit busca_binaria(array, "animais")
v = busca_binaria(array, "jamais")
print(v)
%timeit busca_binaria(array, "jamais")
```
